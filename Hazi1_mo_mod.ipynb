Dolgozzunk a következő függvénnyel:
$$
f(x) = \frac{x^2 - 3x + 4}{x^2 - 1}
$$


```python
import numpy as np
import matplotlib.pyplot as plt

L = 1 # analitikusan számított határérték
a_n = []
n_values = [0,1]
hits = [0,0]

last_zero_index = None

for i in range(2, 1001):  # Loop 2-től 1001-ig, hogy ne osszunk 0-val
    n = i
    a_n_value = (i**2 - 3*i + 4) / (i**2 - 1)
    a_n.append(a_n_value)
    epsilon = np.random.rand() # 0 és 1 között bármekkora epsilon, nagyon nagyon kicsi is lehet
    
    hit = 1 if (L - a_n_value) <= epsilon else 0  # Vagy bene van, vagy nincs benne

    if hit == 0:
        last_zero_index = n  # Utolsó sikertelen lövés indexe
    
    n_values.append(n)
    hits.append(hit)

# Plot
plt.figure(figsize=(12, 6))
plt.scatter(n_values, hits, marker='o', s=10)  # Kissebb gombócok (s=10)
plt.xlabel("n")
plt.ylabel("hit")
plt.title("Hits vs n")
plt.grid()

# Utoló sikertelen lövés
if last_zero_index is not None:
    plt.annotate(f'Last 0 at n={last_zero_index}', 
                 xy=(last_zero_index, 0), 
                 xytext=(last_zero_index + 20, 0.2), 
                 arrowprops=dict(facecolor='red', arrowstyle='->'),
                 fontsize=12, color='red')
    
plt.show()

```
